```python

from transformers import BertModel,AutoTokenizer
# Download pytorch model


tokenizer = AutoTokenizer.from_pretrained("sinhala-nlp/mbert-olid-en")
```

```python
unqiue_word=set()
for sentence in df['pre_text']:
  words=sentence.split()
  for word in words:
    unqiue_word.add(word)

vocab=tokenizer.get_vocab()

yes_count=0
no_count=0
for word in unqiue_word:
  if (word in vocab):
    yes_count+=1
  else:
    no_count+=1

print(yes_count)
print(no_count)
```

```python
subword_tokens_count = 0

for sentence in df['pre_text']:
    tokens = tokenizer.tokenize(sentence)
    subword_tokens_count += len(tokens) - len(sentence.split())

average_subword_tokens_per_word = subword_tokens_count / len(unique_word)
print(average_subword_tokens_per_word)
```
